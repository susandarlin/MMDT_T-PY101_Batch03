# Ex 4: Daily Maths

You are working in a retail shop. Write a program that computes the total cost for a customer that bought 3 differnt types of fruits : apple, orange, mango. Ask the user to enter the quantity and price for each type.

```python
apple_cost = input("Enter the amount of Apple Price: ")
apple_qty = input("Enter the amount of Apple Qty: ")

orange_cost = input("Enter the amount in Orange Price: ")
orange_qty = input("Enter the amount in Orange Qty: ")

mango_cost = input("Enter the amount in Mango Price: ")
mango_qty = input("Enter the amount in Mango Qty")

total_cost = int(apple_cost) * int(apple_qty) + int(orange_cost) * int(orange_qty) + int(mango_cost) * int(mango_qty)
print("Total Cost is {}".format(total_cost))
```

# Ex 9: Computing

Write a program that asks the user for a number of seconds and prints out how many minutes and seconds that is. For instance, 200 seconds is 3 minutes and 20 seconds.

```python
input_seconds = input("Enter seconds value :")
i_input_seconds = int(input_seconds)

mins = i_input_seconds // 60
seconds = i_input_seconds % 60

print("{} seconds is {} minutes and {} seconds.".format(input_seconds, mins, seconds))
```

# Ex 10: Using Math Module

Write a program that asks the user for a number and then prints out the sine, cosine, and tangent of that number.

```python
import math
num = float(input("Enter a number:"))
sine_value = math.sin(num)
cosine_value = math.cos(num)
tangent_value = math.tan(num)
print("Sine : {}".format(sine_value))
print("Consine : {}".format(cosine_value))
print("Tangent : {}".format(tangent_value))
```
