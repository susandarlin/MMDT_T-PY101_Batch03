```python
#Q5) The following code uses stub functions to design the game. Complete the code.
import numpy as np
def generate_random_num():
    """
    Generates a random integer between 1 and 100 (inclusive) for the user to guess.

    Arg: No input argument
    return: integer: The randomly generated number.
    """
    random_num = np.random.randint(1, 101)
    return random_num

def get_user_input():
    """
    This function asks the user to guess the number. Make sure that the number entered is an integer.
    Arg: No input argument
    return: the number.
    """
    while True:
        try:
            guess_num = int(input("Guess a number between 1 and 100: "))
            return guess_num
        except ValueError:
            print("That's not a valid integer. Please try again.")


def check_the_number(actual_num, guess_num):
    """
    This function checks if the actual number and guess number are the same.
    Arg: actual_num(integer): the number generated by random function
         guess_num (integer): the number user has guessed

    return: boolean : True if the numbers are the same. Else, False.
    """
    if(actual_num == guess_num):
      return True
    return False

def display_the_feedback(is_correct, chances):
    """
    This is a void function to print the feedback. If arg is True, display "bingo".
    Else, display "wrong guess" and indicates the number of chances left.

    Arg: is_correct(boolean): indicates if the guess is correct or not,
         chances (integer): indicates the number of chances left
    """
    if(is_correct == True):
      print("bingo!")
    else:
      print("wrong guess!", "the number of chances left ", chances)

### Do not change the functions below ###
def guess_num_game():
    actual_num = generate_random_num()

    max_chances = 10
    for counter in range(1, max_chances+1):
        guess_num = get_user_input()
        is_correct = check_the_number(actual_num, guess_num)
        display_the_feedback(is_correct, max_chances - counter)
        if(is_correct):
            return "You did it! You won the game"

    return "You lost! Play one more time? "


def main():

    while True:
        win_lost_msg = guess_num_game()
        print(win_lost_msg)
        is_continue = int(input("Enter 1 to continue:"))
        if(not is_continue):
            return "Bye. Thanks for playing."

main()
```
